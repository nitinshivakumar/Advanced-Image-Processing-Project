```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import argparse
import json
import array as arr
import os

path = "data/images_panaroma/"
dummy = []
for i in os.listdir(path):
    path1 = path + "/" + i
    img = cv2.imread(path1)
    dummy.append(img)
```

```python
def extract_descriptor_features_function(image):
    shift_detect = cv2.SIFT_create()
    (keypoints, features) = shift_detect.detectAndCompute(image, None)
    return (keypoints, features)
    
def feature_matching(features1, features2):
    matched = []
    for i, j in enumerate(features1):
        distances = np.linalg.norm(features2 - j, axis=1)
        sorted_val = np.argsort(distances)
        output1 = sorted_val[0]
        output2 = sorted_val[1]
        if distances[output1] < 0.75 * distances[output2]:
            matched.append((i, output1, distances[output1]))

    return matched

def homography_stitching(keypoints1,keypoints2,threshold, matched_features):   

    image1_keypoints = np.float32([i.pt for i in keypoints1])
    image2_keypoints = np.float32([i.pt for i in keypoints2])
    
    if len(matched_features) > 4:

        image1_cordinate = np.float32([image1_keypoints[i[0]] for i in matched_features]).reshape(-1, 1, 2)
        image2_cordinate = np.float32([image2_keypoints[i[1]] for i in matched_features]).reshape(-1, 1, 2)
        
        
        Homography_Matrix, status = cv2.findHomography(image1_cordinate, image2_cordinate, cv2.RANSAC, threshold)

        return Homography_Matrix, status
    else:
        None

def stitch_images(image1, image2, Homography_Matrix,ratio=0.75, showMatches=False):
    width =image2.shape[1] + image1.shape[1]

    height = max(image2.shape[0], image1.shape[0])

    result = cv2.warpPerspective(image2, Homography_Matrix, (width, height))
    result[0:image1.shape[0], 0:image1.shape[1]] = image1
   
    return result

def eliminate_black_padding(stitched_image):
    if len(stitched_image.shape) == 3:
        grayscale_image = cv2.cvtColor(stitched_image, cv2.COLOR_BGR2GRAY)
    else:
        grayscale_image = stitched_image

    left_boundary = np.argmax(np.any(grayscale_image > 0, axis=0))
    right_boundary = grayscale_image.shape[1] - np.argmax(np.any(grayscale_image > 0, axis=0)[::-1])

    cropped_image = stitched_image[:, left_boundary:right_boundary]

    return cropped_image

    
```

```python
def best_image_pair(images):
    num_images = len(images)
    final_pair = []

    for i in range(len(images)):
        best_score = float('-inf')
        best_pair = None
        i = 0
        for j in range(i + 1, len(images)):
            image1 = images[i]
            image2 = images[j]

            (keypoints1, features1) = extract_descriptor_features_function(image1)
            (keypoints2, features2) = extract_descriptor_features_function(image2)
            matched_features1 = feature_matching(features1, features2)

            (Homography_Matrix, status) = homography_stitching(keypoints1, keypoints2, 4, matched_features1)
            score = np.sum(status)

            #----------------------------------------------------------------------
            
            matched_features2 = feature_matching(features2, features1)
            (Homography_Matrix_reverse, status_reverse) = homography_stitching(keypoints2, keypoints1, 4, matched_features2)
            score_reverse = np.sum(status_reverse)

            if max(score, score_reverse) > best_score:
                if score > score_reverse:
                    best_score = score
                    best_pair = (i, j, Homography_Matrix)
                    stitch_img = stitch_images(images[j], images[i], Homography_Matrix)
                    a = j
                else:
                    best_score = score_reverse
                    best_pair = (j, i, Homography_Matrix_reverse)
                    stitch_img = stitch_images(images[i], images[j], Homography_Matrix_reverse)
                    a = j

        # plt.imshow(images[i])
        plt.imshow(images[i])
        stitch_img = eliminate_black_padding(stitch_img)         
        if len(images) > 2:
            images.pop(i)
            images.pop(a-1)
            images.insert(0, stitch_img)
    plt.figure(figsize=(20,10))
    plt.imshow(stitch_img)
    return stitch_img
```

```python
path = "data/images_panaroma/"
dummy = []
for i in os.listdir(path):
    path1 = path + "/" + i
    img = cv2.imread(path1)
    dummy.append(img)


best_image_pair(dummy)
```
